```python
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def error(y_head,y):
    return np.sum(0.5*(y_head-y)**2)
```

```python
def initialize_network(no_of_layers, layer_sizes):
    if no_of_layers!=len(layer_sizes):
        return "Please enter valid info"
    W = []
    for i in range(no_of_layers-1):
        _current = layer_sizes[i]
        _next = layer_sizes[i+1]
        w = np.random.normal(0, 1, _current*_next).reshape(_next,_current)
        W.append(w)
    
    return W
```

```python
def forward_prop(x,W):
    a = x.copy()
    layer_outputs = [a.reshape((a.shape[0],1))]
    for w in W:
        z = w.dot(a)
        a = sigmoid(z)
        a = a.reshape((a.shape[0],1))
        layer_outputs.append(a)
    
    return layer_outputs
```

```python
def back_prop(X,y,W,learning_rate=0.1,no_of_epochs=10000):
    ERR = []
    for _ in range(no_of_epochs):
        gradients = [np.zeros(w.shape) for w in W]
        total_error = 0
        for idx, x in enumerate(X):
            layer_outputs = forward_prop(x,W)
            i = -1
        
            y_current = y[idx].reshape((y[idx].shape[0],1))
            deltas = [layer_outputs[i] - y_current]
            total_error += float(error(layer_outputs[i],y_current))
            for w in reversed(W[1:]):
                i = i-1
                current = layer_outputs[i]
                layer_delta = w.T.dot(deltas[0])*current*(1-current)
                deltas.insert(0, layer_delta)
                
            for i in range(len(gradients)):
                gradients[i] = gradients[i] + deltas[i].dot(layer_outputs[i].T)


        ERR.append(total_error)
        for i in range(len(W)):
            W[i] = W[i] - learning_rate*(1/len(X))*gradients[i]
            
    return W, ERR
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()

X = iris['data']
y = iris['target']

temp = np.zeros((len(y),3))
for i in range(temp.shape[0]):
    temp[i][y[i]]=1
    
y = temp
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/6)

W = initialize_network(4, [4,10,10,3])
```

```python
W_end, ERR = back_prop(X_train,y_train,W,learning_rate=0.1,no_of_epochs=10000)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure()
plt.plot(np.arange(len(ERR)),ERR)
plt.xlabel('Epoch')
plt.ylabel('Total Cost')
plt.show()
```

```python
y_score = []
for sample in X_test:
    result = forward_prop(sample, W_end)
    max_idx = np.argmax(result[-1].T)
    y_score.append(max_idx)
```
